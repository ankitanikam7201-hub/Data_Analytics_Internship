Customer Behavior Analysis

To analyze customer purchasing patterns based on age, gender, and product categories.

```python
# Import libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load dataset
df = pd.read_csv("customer_data.csv")
```

```python
# Display dataset
print("First 5 rows:")
df.head()
```

```python
# Data information
print("\nDataset Info:")
df.info()
```

```python
# Check missing values
print("\nMissing Values:")
print(df.isnull().sum())
```

```python
# Data cleaning
df = df.dropna()
```

```python
# Basic statistics
print("\nStatistical Summary:")
df.describe()
```

```python
# Age vs Purchase
df.groupby('age')['purchase_amount'].mean().plot(kind='bar')
plt.title("Average Purchase by Age")
plt.xlabel("age")
plt.ylabel("purchase_amount")
plt.show()
```

```python
# Gender vs Purchase
plt.figure()
sns.barplot(x='gender', y='purchase_amount', data=df)
plt.title("Gender vs Purchase")
plt.show()
```

```python
# Product Category
plt.figure()
df['product_category'].value_counts().plot(kind='bar')
plt.title("Product Category Distribution")
plt.show()
```

```python
# Payment Method
plt.figure()
df['purchase_frequency'].value_counts().plot(kind='bar')
plt.title("Payment Method Usage")
plt.show()
```

```python
# Top customers
print("\nTop Customers:")
print(df.groupby('id')['purchase_amount']
      .sum().sort_values(ascending=False).head())
```

```python
# Correlation heatmap
plt.figure()
sns.heatmap(df.corr(numeric_only=True), annot=True)
plt.title("Correlation Heatmap")
plt.show()
```

- Young customers tend to purchase more products
- Female/Male customers contribute significantly to sales
- Some product categories generate higher revenue
- A few customers contribute to most of the sales
