Website Traffic Analysis

To analyze website traffic patterns, user behavior, and traffic sources to improve engagement and performance.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load dataset
df = pd.read_csv("website_data.csv")
```

```python
# Display data
df.head()
```

```python
df.info()
```

```python
# Missing values
print(df.isnull().sum())
```

```python
# Cleaning
df = df.dropna()
```

```python
# Traffic source
plt.figure()
df['Traffic Source'].value_counts().plot(kind='bar')
plt.title("Traffic Source")
plt.show()
```

```python
avg_page_views = df['Page Views'].mean()
avg_session = df['Session Duration'].mean()
avg_bounce = df['Bounce Rate'].mean()

print("Average Page Views:", avg_page_views)
print("Average Session Duration:", avg_session)
print("Average Bounce Rate:", avg_bounce)
```

```python
# Page views distribution
plt.figure()
df['Page Views'].hist()
plt.title("Page Views Distribution")
plt.show()
```

```python
# Session vs conversion
plt.figure()
sns.scatterplot(x='Session Duration', y='Conversion Rate', data=df)
plt.title("Session vs Conversion")
plt.show()
```

```python
# Bounce rate trend
plt.figure()
df['Bounce Rate'].plot(kind='line')
plt.title("Bounce Rate Trend")
plt.show()
```

```python
# Correlation heatmap
sns.heatmap(df.corr(numeric_only=True), annot=True)
plt.title("Correlation Heatmap")
plt.show()
```


- Some traffic sources bring more visitors than others
- Higher session duration leads to better conversion
- High bounce rate reduces performance
- Page views vary across users
- Website performance depends on user engagement
